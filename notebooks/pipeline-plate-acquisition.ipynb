This notebook sets up and executes a pipeline plate acquisition. There is very little user input required; the only cell that should be edited is the one in which the directory to which the images and logfiles should be saved.

```python
import os
import sys
from py4j.java_gateway import JavaGateway

sys.path.insert(0, '..')
from dragonfly_automation import utils
from dragonfly_automation.gateway import gateway_utils
from dragonfly_automation.fov_models import PipelineFOVScorer
from dragonfly_automation.acquisitions.pipeline_plate_acquisition import PipelinePlateAcquisition
```

### Define the metadata for the plate being imaged

Here we specify the ID of the experiment, the ID of the pipeline plate being imaged, and which half of the plate is being imaged.

__PML ID__<br>
The PML ID is the ID of the imaging experiment/acquisition. It is a string of the form `'PML<number>'` where number is a zero-padded four-digit experiment number. This number must first be manually defined in the 'Pipeline-microscopy-master-key' Google sheet and then copied here.

__Plate ID__<br>
The plate_id parameter must be a string of the form `'P<number>'` where number is a zero-padded four-digit plate number. For example, the plate_id of 'Plate1' is `'P0001'` and the plate_id of 'Plate19' is `'P0019'`. 

__Platemap type__<br>
The platemap_type must be either `'first-half'`, `'second-half'`, or `'custom'`. 

```python
pml_id = ''
plate_id = ''
platemap_type = ''
```

### Load the trained FOV-scoring class

This cells should be run without modification. It takes a few seconds to run and prints some output that can be ignored.

```python
fov_scorer = PipelineFOVScorer(mode='prediction')
fov_scorer.load('../models/2019-10-08/')
fov_scorer.train()
fov_scorer.validate()
```

### Initialize the acquisition

None of the parameters here should need to be changed.

```python
data_dir = os.path.join('D:', 'PipelineML', 'data')
aq = PipelinePlateAcquisition(
    data_dir, 
    fov_scorer, 
    env='prod', 
    pml_id=pml_id,
    plate_id=plate_id,
    platemap_type=platemap_type)

aq.setup()
```

### Start the acquisition

This cell runs the acquisition itself, which generally takes 4-12 hours to complete. If everything is working, log messages should start to appear here within a few seconds. They should be self-explanatory. Note that some log messages may be error messages, which is normal-ish and does not indicate a problem. However, if a real uncaught Python exception occurs (that is, lots of unreadable and color-coded text suddenly appears and the log message stop appearing), then something unexpected and bad has happened, and Keith should be contacted.

Note that AFC timeout errors will __not__ cause the script to crash but will generate a logged error message.

```python
aq.run()
```
